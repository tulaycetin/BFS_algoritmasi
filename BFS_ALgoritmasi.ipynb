# BFS (Breadth-First Search)
Bir graf üzerinde arama veya gezme işlemi yapan temel bir algoritmadır.
Bir başlangıç noktasından başlayarak önce o noktaya komşu olan tüm düğümleri (noktaları) ziyaret eder, ardından bu komşuların komşularına geçer.
Yani, grafı katman katman (seviye seviye) dolaşır.

# Çalışma Mantığı

BFS algoritması kuyruk (queue) veri yapısını kullanır.
Her adımda



1.   Başlangıç düğümünden başlanır ve ziyaret edilir

1.   Bu düğümün komşuları kuyruğa eklenir.

2.   Kuyruğun başındaki düğüm alınır, işlenir ve onun komşuları kuyruğa eklenir.

2.   Kuyruk boşalana kadar bu işlem devam eder.



```python
from collections import deque  # Kuyruk veri yapısı için deque sınıfını içe aktarıyoruz


def bfs(graph, start):
    visited = set()             # Ziyaret edilen düğümleri tutmak için küme
    queue = deque([start])      # Başlangıç düğümünü içeren bir kuyruk oluşturuyoruz
    visited.add(start)          # Başlangıç düğümünü ziyaret edilmiş olarak işaretliyoruz

    while queue:                # Kuyruk boş olmadığı sürece çalış
        node = queue.popleft()  # Kuyruğun başındaki düğümü çıkar
        print(node,"-")             # Ziyaret edilen düğümü ekrana yazdır

        # Şimdiki düğümün tüm komşularını dolaş
        for neighbor in graph[node]:
            # Eğer bu komşu daha önce ziyaret edilmemişse
            if neighbor not in visited:
                visited.add(neighbor)   # Ziyaret edildi olarak işaretle
                queue.append(neighbor)  # Kuyruğun sonuna ekle (bir sonraki seviyede işlenecek)
```

Bu garf yönlü  bir graftır. Ok yönüne göre her düğümün hangi düğüme gittiğini çıkararak komşuluk listesi  bululurz
örneğin
**A → B, A → C, A → D, B → E, E → H, H → J gibi.**

```python
# Örnek bir graf tanımlayalım (komşuluk listesi şeklinde)
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': ['H'],
    'F': ['H'],
    'G': ['H'],
    'H': ['I', 'J'],
    'I': ['J'],
    'J': ['K'],
    'K': []
}

# BFS algoritmasını 'A' düğümünden başlatarak çalıştır
bfs(graph, 'A')
```
